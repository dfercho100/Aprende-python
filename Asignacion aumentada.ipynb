La asignacion aumentada es aumentarle o restarle a un valor, es decir es sumar o restar a una variable, esto se hace con el aperador += para sumarle o -= para restarle (con el ejemplo va a quedAar mas claro), la sintaxis de este operador es: variable menor que operator mayor que = value O la forma mas eficiente de hacerlo es: variale = variable menor que operator mayor que value

Vamos a hacer uso de este operador aumentandole 5 a la variable existente:

```python
my_var = 10
my_var += 5

print(my_var) # 15
```

que es lo mismo que escribir:

```python
my_var = 10
my_var = my_var+ 5

print(my_var) # 15
```

La ventaja de este operador es que puede actualizar rapidamente el valor de una variable sin tener que repetir la wl nombre de la variable, reduciendo redundancias y posibles errores. 

Vamos a reducir le a una variable:

```python
count = 14
count -= 3

print(count) # 11
```

Vamos a multiplicarle a una variable, esto se hace con el operador *= :

```python
product = 65 
product *= 7
print(product)
```

Vamos a dividir el precio de una variable:

```python
price=30
price /= 2
print(price)
```

En python a otro operador que se denomina dision de piso y este se representa como // se refiere generalmente al uso de la operación de división entera, la cual devuelve el cociente de una división redondeado hacia abajo al entero más cercano. y asi es la division de piso:

```python
resultado = 7 // 3
print(resultado)
```

En este caso, 7 // 3 devuelve 2, ya que la división de 7 entre 3 da 2.33, y con la división de piso, se redondea hacia abajo.

```python
print(10 // 4)  # Imprime 2
print(-10 // 4)  # Imprime -3
print(7 // 3)    # Imprime 2
print(7 // -3)   # Imprime -3
```

Volviendo al operador de asignacion aumentada, vamos a hacer una division de piso a la variable:

```python
total_pages = 23
total_pages //= 5

print(total_pages) 
```

Vamos a hacer una asignacion de modulo calculando el resto de la variable:

```python
bits=35
bits %= 2
print(bits)
```

la asignacion de exponenciasion **= eleva la variable a la potencia:

```python
power=2
power **= 3
print(power)
```
