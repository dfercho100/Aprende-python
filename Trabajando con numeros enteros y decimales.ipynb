# Agregando funciones 

En python puede agrefar funciones para alterar, modificar entre otras y una de estas funciones es la funcion float() 

la funcion float() puede convertir un entero a un numero decimal añadiendo un punto decimal 0 al entero 

```python
my_int_1 = 56
my_float_4=float(my_int_1)
print('convertir int a float:', my_float_4)
```

Tambien puede hacer lo mismo de float a int con la fincion int():

```python
my_float_4=4.32423454
my_int_4=int(my_float_4)
print('convertir float a int:', my_int_4)
```

Tambien puede usar las mismas funciones convirtiendo un string a un enteero o a un float:

```python
my_str_int = "123"
my_str_float = "4.56"

converted_int = int(my_str_int)
converted_float = float(my_str_float)
print('convertir str a int:', converted_int)
print('convertir str a float:', converted_float)
```

La funcion round() redondea los numeros decimales:

```python
my_int_4 = 7.233232
my_int_5 = 2.98765
round(my_int_4)
round(my_int_5)

print('redondear:', round(my_int_4))
print('redondear:', round(my_int_5,1))
```

bin() convierte un entero a un  binario:

```python
number = 98

abs(number)
print('valor absoluto:', abs(number))
```

oct() convierte un  entero a un octal: 

```python
my_int_1=56

bin(my_int_1)
print('binario de 56:', bin(my_int_1))
```

hex() convierte un entero a un hexagesimal:

```python
hex(my_int_1)
print('hexadecimal de 56:', hex(my_int_1))
```

pow() eleva un  numero a la potencia de otro o realiza la exponenciacion modular:

```python
pow(my_int_3, my_int_2) # equivalente a my_int_3**my_int_2
print('potencia de int:', pow(my_int_3, my_int_2))

result_1= pow(2,3,5)# equivalente a (2**3)%5
print('potencia con modulo:', result_1)
```
