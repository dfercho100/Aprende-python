# LISTAS: Que son y como los trabajo?

Cuando se habla de listas, trabajamos con listas, tuplas y rangos, estoas son los tres tipos de secuencias mas usados en python. 
Los tipos de listas son una secuencia orrdenada de datos que pueden estan compuestas por cadenas de texto, numeros u otros, las listas son mutables y usan cero basado en index lo que quiee decir que su indice es cero; la sintaxis basica de una lista es:
ciudades['Los Angeles', 'London', 'Tokyo']

```python
cities = ['Los Angeles', 'London', 'Tokyo']
cities[0]
```

el inddice negativo es el ultimo de la lista, que en este caso es Tokyo:

```python
cities = ['Los Angeles', 'London', 'Tokyo']
cities[-1] 
```

Hay otro camino para usar una lista que es usando el constructor list() y se usaq de la siguiente manera:

```python
developer = 'Jessica'
list(developer)
```

```python
homework = 'Python' 'java' 'c++'
list(homework)
```

como vemos usando este constructor se imprime letra a letra el valor 

Listo, ahora para obtener el numero total de una lista de numeros se puede usar la funcion len():

```python
numbers = [1, 2, 3, 4, 5]
len(numbers)
```

Si lo que se quiere  es eliminar un elemento de unna lista se usa la palabra clave del:

```python
developer = ['Jane Doe', 23, 'Python Developer']
del developer[1]
print(developer)
```

A veces es conveniente saber si un elemento esta dentro de una lista 

```python
'London' in cities
```

```python
'HTML' in homework
```

Si nos encontramos con listas anidadas, accedemos al indice desde donde empieza esa lista anidada:

```python
developer = ['Alice', 25, ['Python', 'Rust', 'C++']]
developer[2]
```

¿que pasa si quiero acceder a al segundo lenguaje de la lista anidada?  pues uso el indice 1:

```python
developer = ['Alice', 25, ['Python', 'Rust', 'C++']]
developer[2][1]
```
