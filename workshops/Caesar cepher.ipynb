En este taller, vas a construir un cifrado C√©sar. Esta es una de las t√©cnicas m√°s sencillas para cifrar texto, que consiste en sustituir cada letra del texto plano por la letra que se encuentra en un n√∫mero fijo de posiciones m√°s abajo en el alfabeto. Por ejemplo, con un desplazamiento de 5, a se reemplazar√≠a por f, b por g, y as√≠ sucesivamente.

Crea una variable llamada Shift y asignale el valor 5, declara otra variable llamada alphabet y asignale el siguiente string: "abcdefghijklmnopqrstuvwxyz". 

```python
Shift = 5
alphabet = "abcdefghijklmnopqrstuvwxyz"
```

Crea una variable llamada shifted_alphabet y usa el slicing sintax(o la sintaxis de corte) para asignarlo en la pocision de alphabet que comienza en el indice Shift; con la funcion print() imprime shifted_alphabet.

```python
shifted_alphabet = alphabet[Shift:] + alphabet[:Shift]
print(shifted_alphabet)
```

El metodo str.maketrans() es utilizado para crear una tabla de traslacion que mapea caracteres de una cadena a otra. es decir permite definir como se deben reemplazar ciertos caracteres en una cadena dada. Cree una tabla de traduccion que mapee los caracteres de alphabet a los caracteres de shifted_alphabet asignele una variable llamada translation_table luego declare otra nueva variable llamada text y asignele el string "Hello word" que sera el mensaje a encriptar.

```python
translation_table = str.maketrans(alphabet, shifted_alphabet)
text = "Hello word"
```

El m√©todo translate() toma como argumento la tabla de traducci√≥n generada por str.maketrans(). Se llama en una cadena y devuelve una copia de la cadena original donde los caracteres han sido reemplazados seg√∫n la tabla de traslaci√≥n.
Llama el metodo translate() en text pasando translation_table como el argumento y asignando el resultado a una variable llamada encrypted_text e imprimelo para ver el resultado.

```python
encrypted_text = text.translate(translation_table)
print(encrypted_text)
```

Crea una funcion llamada caesar, pon todo el codigo desarrollado hasta ahora dentro del cuerpo de la funcion y ten cuidado con  la identacion

```python
def caesar():
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    shift = 5
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    translation_table = str.maketrans(alphabet, shifted_alphabet)
    text = "Hello word"
    encrypted_text = text.translate(translation_table)
    print(encrypted_text)

# Prueba la funci√≥n
result = caesar()
print(result)
```

Dale a la funcion caesar dos parametros nombrados text y shift y eliminalos del curpo de la funcion. Prueba la funcion caesar llamandolo con el string PythonCode y asignale el numero 3 como argumento, asigna la funcion llamada a la variable encrypted_text 

```python
def caesar(text, shift):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    translation_table = str.maketrans(alphabet, shifted_alphabet)
    encrypted_text = text.translate(translation_table)
    return encrypted_text # se elimina el print para retornar el valor

encrypted_text = caesar("PythonCode", 3)
print(encrypted_text)
```

Al notar el resultado que imprimio los caracteres en mayuscula han quedado intactos lo cual ocurre porque la tabla traduccion no incluye letras mayusculas, actualiza la llamada str.maketrans() concatenando a cadda argumento la version mayuscula del propio argumento

```python
def caesar(text, shift):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    translation_table = str.maketrans(alphabet + alphabet.upper(), shifted_alphabet + shifted_alphabet.upper())
    encrypted_text = text.translate(translation_table)
    return encrypted_text

encrypted_text = caesar("PythonCode", 3)
print(encrypted_text)
```

Usa ell operador not en la sentencia if. Crea otro if que verifique que shift es menor que 1 y asi retornar el string "Shift debe ser un entero entre 1 y 25", en el segundo if verifica que shift sea mayor que 25 utilizando el operador logico or 

```python
def caesar(text, shift):
    if not isinstance(shift, int): #isinstance verifica el tipo de dato
        return "Shift necesita ser un valor entero"
    
    if shift < 1 or shift > 25:
        return "Shift debe ser un entero entre 1 y 25"
    
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    translation_table = str.maketrans(alphabet + alphabet.upper(), shifted_alphabet + shifted_alphabet.upper())
    encrypted_text = text.translate(translation_table)
    return encrypted_text

encrypted_text = caesar("PythonCode", 3)
print(encrypted_text)
```

A√±ade un tercer parametro que se llame encrypt a tu funcion caesar y dale un valor por defecto de verdadero, para determinar si la funci√≥n debe cifrar el texto que le pasa (comportamiento por defecto, encrypt=True), o si debe descifrar un mensaje cifrado.

Crea una sentencia if que compruebe si encrypt no es truthy. Dentro de la nueva instrucci√≥n if, configura shift a - shift. Esto es necesario para permitir que el cambio se produzca en la direcci√≥n opuesta respecto al proceso de cifrado.

```python
def caesar(text, shift, encrypt=True):
    if not isinstance(shift, int):
        return "Shift necesita ser un valor entero"
    
    if shift < 1 or shift > 25:
        return "Shift debe ser un entero entre 1 y 25"
    
    if not encrypt:
        shift = -shift
    
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    shifted_alphabet = alphabet[shift:] + alphabet[:shift]
    translation_table = str.maketrans(alphabet + alphabet.upper(), shifted_alphabet + shifted_alphabet.upper())
    encrypted_text = text.translate(translation_table)
    return encrypted_text

# Prueba la funci√≥n
# aqui esroy cifrando el texto
encrypted_text = caesar("PythonCode", 3)
print(encrypted_text)

# Prueba descifrando
# aque estoy descifrando el texto
decrypted_text = caesar("SbwkrqFrgh", 3, encrypt=False)
print(decrypted_text)
```
