no es posible concatenar un nombre con un numero ya que al hacerlo genera un TypeError,  si lo que se quiere es concatenar un sting con un numero convierta el nummero en una cadena con la funcion str()

```python
name = 'John Doe'
age = 26

name_and_age = name + str(age)
print(name_and_age) # John Doe26
```

otra forma de concatenar string con int, use += y realiza  la conncatenacion y la asignacion en un solo paso 

```python
name = 'John Doe'
age = 26

name_and_age = name  # Start with the name
name_and_age += str(age)  # Append the age as string

print(name_and_age)  # John Doe26
```

Aparte de los strings regulares, Pytthon tiene una categoria de string llamada f-strings que permite manejar la interpolacion y hacer alggo de concatenacion con una sintaxis compacta y legible 

Los f-strings comienzan con f (mauscula o minuscula) antes de las comillas y permites incrustar variiables u expresiones dentro de llaves{}

```python
name = 'John Doe'
age = 26
name_and_age = f'My name is {name} and I am {age} years old'
print(name_and_age) # My name is John Doe and I am 26 years old

num1 = 5
num2 = 10
print(f'The sum of {num1} and {num2} is {num1 + num2}') # The sum of 5 and 10 is 15
```

¡Ya sabemos manejar los strings! ahora Podemos hacer muchas cosas con ellos como por ejemplo saber la longitud de los strings con la funcion len()

```python
my_str = 'Hello world'
print(len(my_str))  # 11
```

....O podemos Indexarlos, Cada posicion en un string tiene una  posiscion llamada index. 

El primer caracter de un string es 0, el index del segundo caracter es 1 y asi.. para acceder a un caracter por su indice us corchetes [] 

```python
my_str = "Hello world"

print(my_str[0])  # H
print(my_str[6])  # w
```

con los index negativos funciona al reves en donde el último caracter es -1, el segundo es -2 y asi.... 

```python
my_str = 'Hello world'
print(my_str[-1])  # d
print(my_str[-2])  # l
```

...¡O se pueden cortar!. con string slicer se puede extraer una poorcion del texto o trabbajar con una porcion especifica del texto; si quiere extraer parte del texto use :(dos puntos).

(la sintaxis de string slicing es 
 * string[start:stop]
)

```python
my_str = 'Hello world'
print(my_str[1:4]) # no incluye el index 4, por lo que imprime los index 1,2,3
```

Puede omitir los indices start y stop, eso ocurre si usted omite el indice start

```python
my_str = 'Hello world'
print(my_str[:7])  # Hello w
```

Esto ocurre si omite el indice stop 

```python
my_str = 'Hello world'
print(my_str[8:])  # rld 
```

un paso adicional es el corte o silice, que se usa para especificar el incremento entre cada indice 
(su sistaxis es:
* string[start:stop:step]
)

vamos a empezar desde el indice 0, paramos antes del indice 11 y se extrae cada tercer caracter 

```python
my_str = 'Hello world'
print(my_str[0:11:3])  # Hlwr
```

se puede invertir el string poniendo -1(o -2 o -3....) en el parametro step:

```python
my_str = 'Hello world'
print(my_str[::-1]) # dlrow olleH
```

# Metodos string mas comunes

metodos que trabajan con strings

* upper() retorno un string y lo onvierte en mayuscula

```python
my_str = 'hello world'

uppercase_my_str = my_str.upper()
print(uppercase_my_str)  # HELLO WORLD
```

* lower(): retorna un string y lo convierte en minuscula 

```python
my_str = 'Hello World'

lowercase_my_str = my_str.lower()
print(lowercase_my_str)  # hello world
```

* strip(): devuelve la cadena eliminando los espacios en lanco delanteros y traseros 

```python
my_str = '  hello world  '

trimmed_my_str = my_str.strip()
print(trimmed_my_str)  # "hello world"
```

* replace (old,new ): retorna un nuevo string reemplazando lo viejo por lo nuevo 

```python
my_str = 'hello world'

replaced_my_str = my_str.replace('hello', 'hi')
print(replaced_my_str)  # hi world
```

* split(separator): si el string esta dentro de comillas, puede usar este sepaarador para DIVIDIRLO 

```python
my_str = 'hello world'

split_words = my_str.split()
print(split_words)  # ['hello', 'world']
```

* join(iterable): al contrario de split(separator), join UNE el string 

```python
my_list = ['hello', 'world']

joined_my_str = ' '.join(my_list)
print(joined_my_str)  # hello world
```

* endswith(suffix): Devuelve un booleano que indica si una cadena termina con el sufijo especificado.

```python
my_str = 'hello world'
starts_with_hello = my_str.startswith('hello')
print(starts_with_hello)  # True

my_str = 'hello world'
starts_with_hello = my_str.startswith('terminal')
print(starts_with_hello)
```

* count(substring): retorna el numero de veces que aparece una subcadena en una cadena 

```python
my_str = 'hello world'

o_count = my_str.count('o')
print(o_count)  # 2
```

 * capitalize(): Devuellve una nueva cadena con el primer caracter en mayuscula y los demas en minusculas 

```python
my_str = 'hello world'

capitalized_my_str = my_str.capitalize()
print(capitalized_my_str)  # Hello world
```

* isupper(): retorna true si todas los caracteres de la cadena estan en mayusccula y false si no 

```python
my_str = 'hello world'

is_all_upper = my_str.isupper()
print(is_all_upper)  # False
```

* islower(): retorna true si todas los caracteres de la cadena estan en minuscula y false si no 

```python
my_str = 'hello world'

is_all_lower = my_str.islower()
print(is_all_lower)  # True
```

* title(): retorna el straing con cada primera letrra de cada palabra en mayuscula

```python
my_str = 'hello world'

title_case_my_str = my_str.title()
print(title_case_my_str)  # Hello World
```
