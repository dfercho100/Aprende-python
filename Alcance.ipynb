El alcance determina el orden en que se busscan las variables,  Existen diferentes tipos de alcance, que determinan en qué partes del programa se pueden acceder a las variables. El acance se clasifica segun el modelo LEGB que se definen en tres ámbitos principales:
* Alcance local 
* Alcance de contencion 
* Alcance global 
* Alcance incorporado 

O por sus siglas en ingles:
* **L**ocal scope 
* **E**nclosing scope 
* **G**lobal scope 
* **B**uilt-in scope 

# Local scope

Son las variables que estan declaradas DENTRO de una funcion o clase.:

```python
def my_funtion():
    my_var=10
    print(my_var)
my_funtion()
```

en este caso la funncion my_function tiene su propio alcance y no se puede acceder fuera de la funcion, Imprime 10 ya que la funcion esta dentro pero...

Si esta fuera de la funcion arrojaria un NameError indicando que my_var no esta definido:

```python
def my_funtion():
    my_var=10
    print(my_var)
my_funtion()
print(my_var)
```

# Encosing scope 

Quiere decir que una funcion que esta anidada dentro de otra funcion puede acceder a las variables de la funcion a la que esta anidada. (o en otras palabras se puede acceder a la funcion que esta FUERA de una funcion o clase):

```python
def outer_func():
    msg = 'Hello there!'

    def inner_func():
        print(msg)

    inner_func()

outer_func() # Hello there!
```

Otro ejemplo de Encoding score:

```python
def outer_func():
    msg = 'Hello there!'
    res = ""  # Declare res in the enclosing scope

    def inner_func():
        nonlocal res  # Allow modification of an enclosing variable
        res = 'How are you?'
        print(msg)  # Accessing msg from outer_func()

    inner_func()
    print(res)  # Now res is accessible and modified

outer_func()
```

# Global scope 

Las variables definidas fuera de cualquier función o clase tienen un alcance global y pueden ser accesibles desde cualquier parte del código. Sin embargo, si una variable global es modificada dentro de una función, es necesario usar la palabra clave global para indicar que se debe hacer referencia a la variable global.

```python
x = 5  # x tiene un alcance global

def mi_funcion():
    global x  # Permite modificar la variable global
    x = 10

mi_funcion()
print(x)  # Imprime 10 porque la variable global fue modificada
```

# Built-in scope 

la Built-in scope son todas las funciones, modulos y palabras clave  integradas en python y se pueden acceder en cualquier lugar de programa:

```python
print(str(45)) # '45'
print(type(3.14)) # <class 'float'>
print(isinstance(3, str)) # False
```
