# Condicionales 

Los condicioneles y los numeros numeros booleanos dan como resultado true o false, los condicionales son: mayor que menor que igual mayor e igual que menor e igual que y no es igual:

```python
print(3 > 4) # False
print(3 < 4) # True
print(3 == 4) # False
print(4 == 4) # True
print(3 != 4) # True
print(3 >= 4) # False
print(3 <= 4) # True
```

# Bucle if else y condicional elif

Podemos agregar un bocle if de la siguiente manera:

```python
age = 18

if age >= 18:
    print('You are an adult') 
    
```

Es como decir si quieres esto, imprimelo

Tambien existe else en phyton que es como decir "si no quiere estos, entonces toma esto":

```python
age = 12

if age >= 18:
    print('You are an adult')
else:
    print('You are not an adult yet')
```

Pero paython añade mas opciones a la conndicional if else, que es elif, que como decir "te doy variable, si es verdad lo que indica en la primera condicion imprimela, pero si la segunda condicion es verdadera y la primera es falsa imprime la segunda.... y si ninguna de las dos es verdadera imprime la tercera opcion": 

```python
age = 12

if age >= 18:
    print('You are an adult')
elif age <= 13:
    print('You are a teenager')
else:
    print('You are a child') 
```

se pueden agregar tantos elif como desee: 

```python
age = 2

if age >= 65:
    print('You are a senior citizen')
elif age >= 30:
    print('You are an adult in your prime')
elif age >= 18:
    print('You are a young adult')
elif age >= 13:
    print('You are a teenager')
elif age >= 3:
    print('You are a young child')
else:
    print('You are a toddler or an infant')
```

# Condicional anidado 

Hay situaciones en los que te enfrentarás en  situaciones en la se tienen que comparar varios valores a la vez y aqui es donde entran las condicionales anidadas:

```python
is_citizen = True
age = 25

if is_citizen:
    if age >= 18:
        print('You are eligible to vote') # You are eligible to vote
else:
    print('You are not eligible to vote')
```

En Python, un if anidado es simplemente una estructura de control condicional donde se coloca un if dentro de otro if. Esto permite realizar verificaciones más complejas.

```python
x = 10
y = 5

if x > y:
    print("x es mayor que y")
    if x > 8:
        print("x es también mayor que 8")
    else:
        print("x es menor o igual a 8")
else:
    print("x no es mayor que y")
```

En este caso:

* Primero se verifica si x es mayor que y.
* Si lo es, se ejecuta un segundo if que evalúa si x es mayor que 8. Si no es mayor que 8, se entra en el bloque else.

Hay 3 opeardores booleanos and or not y cada uno tiejne funciones muy diferentes cuando se emplean

El operador and toma dos operandos y devuelve el primer operando si es falso; de lo contrario devuelve el segundo operando, ambos operandos deben  ser verdaderos para que una expresion dé como resultado un valor verdadero 

```python
print(is_citizen and age)
# da como resultado: 25 porque ambos valores son verdaderos 
```

en phython algunos valores se consideran truthy y falsy. Los valores truthy se consideran como verdaderos y los valores falsy se consideran como falsos. para comprobar si un valor es truthy o falsy se puede usar la funcion bool():

```python
print(bool(False)) # False
print(bool(0))  # False
print(bool('')) # False

print(bool(True)) # True
print(bool(1)) # True
print(bool('Hello')) # True
```
