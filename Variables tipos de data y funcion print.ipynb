# Función print 

Para usar la funcion print, se pone la palabra clave print seguido del mensaje, este mensaje va dentro de comillas simples. Por ejemplo 

```python
print("Nombre:", name)#Nombre:, name se llama argumento
print("Edad:", age)
print("otro nombre:", my_variable_name)
print("edad del otro usuario:", user_age)
```

puede usar print mostrando multiples argumentos separados con comas:

```python
print('Mis colores favoritos son: rojo, azul y verde.')
```

# tipos de datos mas comunes

Python sabe que tipo de datos es una variable segun lo que se le asigne:

```python
name="Diego" #pytohon conoce que esto es una variable de tipo string
age=32#python conoce que esto es una variable de tipo entero
my_variable_name="John Doe" #python conoce que esto es una variable de tipo string
user_age=28 #python conoce que esto es una variable de tipo entero
```

```python
#las imprimimos de nuevo 
print("Nombre:", name)#Nombre:, name se llama argumento
print("Edad:", age)
print("otro nombre:", my_variable_name)
print("edad del otro usuario:", user_age)
```

La siguiente es una lista de datos mas comunes enn Python:

Integer: un numero entero sin decimales 

```python
my_integer_var=10
print("El valor de la variable entera es:", my_integer_var)
```

Float: un numero decimal 

```python
my_float_var=3.14
print("El valor de la variable flotante es:", my_float_var)
```

Complex: un numero con una parte real e imaginaria 

```python
my_complex_var=2 + 3j
print("El valor de la variable compleja es:", my_complex_var)
```

String: una secuencia de caracteres que van entre comillas simples o comillas dobles 

```python
my_string_var="Hello, World!"
print("El valor de la variable string es:", my_string_var)
```

Boolean: un tipo de dato que puede ser true o false 


```python
my_boolean_var=True
print("El valor de la variable booleana es:", my_boolean_var)
```

Set: una colección NO ORDENADA de elementos únicos

```python
my_set_var={1, 2, 3}
print("El valor de la variable conjunto es:", my_set_var)
```

Dictionary: una colección de pares clave-valor encerrados en llaves 

```python
my_dict_var={"name": "Alice", "age": 30}
print("El valor de la variable diccionario es:", my_dict_var)
```

Tuple: una colección ORDENADA e inmutable entre parentesis

```python
my_tuple_var=(1, 2, 3)
print("El valor de la variable tupla es:", my_tuple_var)
```

Range: una secuencia de numeros a menudo usada en bocles 

```python
my_range_var=range(5)
print("El valor de la variable rango es:", list(my_range_var))
```

List: una colección de elementos que soportan diferentes tipos de datos 

```python
my_list_var=[1, 'hello word', 3.14, 4, False]
print("El valor de la variable lista es:", my_list_var)
```

None: un valor especial que representa la ausencia de un valor 

```python
my_none_var=None
print("El valor de la variable None es:", my_none_var)
```
